```python
# Use this cell to begin your analysis, and add as many as you would like!
#Importing Pandas Library
import pandas as pd
```

```python
#Reading CSV file
apps = pd.read_csv("datasets/apps.csv")
apps.head()
```

```python
#Show details about (Installs) column
print("Number of unique elements : {}".format(apps["Installs"].nunique()))
print()
print("--------------------------------------")
print()
apps.info()
print()
print("--------------------------------------")
print()
apps["Installs"].value_counts()
```

```python
#Modifying each element in the Installs column to be ready for being an integer
def modify(x):
    if x[-1] == '+':
        x = x[:-1]
    return int("".join(x.split(',')))

#Modifying the Installs column
apps["Installs"] = apps["Installs"].apply(modify)
```

```python
#Check the Installs column after being modified

#Show details about (Installs) column
print("Number of unique elements : {}".format(apps["Installs"].nunique()))
print()
print("--------------------------------------")
print()
apps.info()
print()
print("--------------------------------------")
print()
apps["Installs"].value_counts()
```

```python
#Finding the number of apps in each category, the average price, and the average rating. 
app_category_info = apps.groupby(["Category"]).agg({'App':'count','Price':'mean','Rating':'mean'})
app_category_info.columns = ['Number of apps','Average price','Average rating']
```

```python
#Reading CSV file
reviews = pd.read_csv("datasets/user_reviews.csv")
reviews.head()
```

```python
#Slicing the records with Finance category from the whole dataframe 
finance_apps = apps[apps["Category"]=="FINANCE"]#Conditional Accessing
finance_apps.head()
```

```python
#Slicing the records with Finance category and Free Type from the finance_apps dataframe 
free_finance_apps = finance_apps[finance_apps["Type"]=="Free"]#Conditional Accessing
free_finance_apps.head()
```

```python
#Merging the free_finance_apps dataframe with reviews dataframe on the App column
new = pd.merge(free_finance_apps,reviews,on='App')
new.head()
```

```python
#Grouping by App column and the aggregation function on the Sentiment Score column
user_feedback = new.groupby("App").agg({"Sentiment Score":'mean'})
user_feedback.head()
```

```python
#Sorting by the Sentiment Score column
user_feedback = user_feedback.sort_values(by=['Sentiment Score'],ascending=False)
user_feedback.head()
```

```python
#Getting the top 10 free FINANCE apps having the highest average sentiment score
top_10_user_feedback = user_feedback[:10]
top_10_user_feedback
```
